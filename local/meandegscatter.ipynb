```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
figdir = '/Users/annabroido/Dropbox/Research/ICONProject/1Scale_Free_Networks/writing/Figures/'
df = pd.read_pickle('/Users/annabroido/Dropbox/Research/LRTAnalysis/LRTAnalysis/analysis/hyps.p')
ns = df.n.values.astype('float')
meandegs = df.median_meandeg.values.astype('float')
```

```python
df.columns
```

```python
ALMOST_BLACK = '0.125'
DARK = '0.4'
GRAY_COLOR = '0.5'
ALL_COLOR = (0.156, 0.254, 0.466)
SUPERWEAK_COLOR = (0.647, 0.772, 0.972)
WEAKEST_COLOR = (0.419, 0.627, 0.960)
WEAK_COLOR = (0.258, 0.525, 0.85)
STRONG_COLOR = (0.223, 0.368, 0.674)
STRONGEST_COLOR = (0.156, 0.254, 0.466)
LABEL_SIZE = 20
TICK_SIZE = 20
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,7))
ax.scatter(np.log10(ns), np.log10(meandegs), s=50, color=DARK, alpha=0.7, edgecolor=ALMOST_BLACK)



divider = make_axes_locatable(ax)
ax2 = divider.append_axes("top", size=1.2, pad=.15, sharex=ax)    
# ax2.hist(np.log10(ns), bins=20, edgecolor=STRONG_COLOR, facecolor='w', histtype='step')
ax2.hist(np.log10(ns), bins=20, color=WEAK_COLOR, edgecolor='w')
ax2.spines["right"].set_visible(False)
ax2.spines["top"].set_visible(False)
# ax2.spines["left"].set_visible(False)
ax2.yaxis.set_ticks_position('left')
ax2.set_yticks([50, 150])
ax2.tick_params(axis=u'both', which=u'both',length=0) 
ax2.tick_params(axis=u'y', which=u'major',labelsize=TICK_SIZE, length=5, pad=5)
ax2.set_ylabel('Number of \nnetworks', fontsize=LABEL_SIZE, labelpad=10)
# ax2.tick_params(axis=u'both', which=u'both',length=0) 

# right histogram
ax3 = divider.append_axes("right", size=1.2, pad=0.2, sharey=ax)
# ax3.hist(np.log10(meandegs), bins=10, edgecolor=WEAKEST_COLOR,facecolor='w', orientation='horizontal', histtype='step')
ax3.hist(np.log10(meandegs), bins=10, facecolor=SUPERWEAK_COLOR,edgecolor='w', orientation='horizontal')
ax3.spines["right"].set_visible(False)
ax3.spines["top"].set_visible(False)
#ax3.spines["bottom"].set_visible(False)
ax3.xaxis.set_ticks_position('bottom')
ax3.set_xlim(0,470)
ax3.set_xticks([200, 450])
ax3.tick_params(axis=u'y', which=u'both',length=0) 
ax3.tick_params(axis=u'x', which=u'major',labelsize=TICK_SIZE, length=5, pad=8)
ax3.set_xlabel('Number of \nnetworks', fontsize=LABEL_SIZE, labelpad=10)


ax.set_xticks([1, 2, 3, 4, 5, 6])
ax.set_xticklabels([r'10$^1$', r'10$^2$', r'10$^3$', r'10$^4$', r'10$^5$',r'10$^6$'])
ax.set_ylim(0,2.05)
ax.set_yticks([1,2])
ax.set_yticklabels([r'10$^1$', r'10$^2$'])
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ylab = r"Mean degree $\langle k \rangle$"
ax.set_ylabel(ylab, fontsize=LABEL_SIZE, labelpad=10)
ax.set_xlabel(r'Number of nodes $n$', fontsize=LABEL_SIZE, labelpad=10)
ax.tick_params(axis='y', which='major', labelsize=TICK_SIZE, pad=5)
ax.tick_params(axis='x', which='both', labelsize=TICK_SIZE+2, pad=5)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.setp(ax2.get_xticklabels(), visible=False)
plt.setp(ax3.get_yticklabels(), visible=False)
plt.savefig(figdir + 'degreedata.png', format='png', bbox_inches='tight', dpi=1000)
plt.show()
```
